## Building class to integrate all the process

```python
from selenium.webdriver.common.keys import Keys
import time
```

```python
class Nifty50:
    
    def __init__(self, driver, url, driver2):
        
        self.driver = driver
        self.driver.get(url)
        
        try:
            view_more = self.driver.find_element(By.LINK_TEXT, 'View More News')
            self.driver.execute_script("arguments[0].click()", view_more)
        
            for i in range(9):
                html = self.driver.find_element(By.TAG_NAME, 'html')
                html.send_keys(Keys.PAGE_DOWN)
                time.sleep(2)
        except:
            pass

        self.news_html = self.driver.find_elements(By.CSS_SELECTOR, "div[class = 'listingNew clearfix']")
    
        self.driver2 = driver2
        
    def article_time(self, i):
        article_time = self.news_html[i].find_element(By.CSS_SELECTOR, "span[class = 'fl date']")
        try:
            return(article_time.text).split('. ')[1]
        except:
            return('')

    def article_title(self, i):
        article_title = self.news_html[i].find_element(By.CSS_SELECTOR, "h2[class = 'headline']")   
        return(article_title.text)

    def article_link(self, i):
        article_link = self.news_html[i].find_element(By.TAG_NAME, 'a')
        try:    
            return(article_link.get_attribute('href'))
        except:
            return('')
    
        
    def article_content(self, article_link):    
        
        try: 
            time.sleep(2)
            self.driver2.get(article_link)
            time.sleep(1)
            
            try:
                main_area = self.driver2.find_element(By.CSS_SELECTOR, "div[class = 'mainArea']")
                text = ' '.join([element.text for element in main_area.find_elements(By.TAG_NAME, "p")])
                return text
            
            except:
                return('')
        except:
            return('')
    
    def scrapping(self):

        news_list = []
        for i in range(len(self.news_html)):
            print(i)
            news_list.append({'article_time': self.article_time(i),
                              'article_title': self.article_title(i),
                              'article_link': self.article_link(i),
                              'article_content': self.article_content(self.article_link(i))
                            })
        
        df = pd.DataFrame(news_list)

        return df
```

```python
def token_frequencies(i):
    
    tokens = word_tokenize(df['article_content'][i])

    token_frequencies = {}

    for token in tokens:
        if token.lower() not in stop_words:
            if token.lower() not in punctuations:
                if token.lower() not in token_frequencies.keys():
                    token_frequencies[token.lower()] = 1
                else:
                    token_frequencies[token.lower()] = token_frequencies[token.lower()] + 1
                    
    return token_frequencies
```

```python
def sent_scores(token_frequencies, i):
    
    sent_tokens = sent_tokenize(df['article_content'][i])
    # print('**',len(sent_tokens))

    sent_scores = {}

    for sent_token in sent_tokens:
        tokens = sent_token.split(' ')
        for token in tokens:
            if token.lower() in token_frequencies.keys():
                if sent_token not in sent_scores.keys():
                    sent_scores[sent_token] = token_frequencies[token.lower()]
                else:
                    sent_scores[sent_token] = sent_scores[sent_token] + token_frequencies[token.lower()]
                    
    return sent_scores
```

```python
def summary(sent_scores, sent_length):

    summary = [key for key, value in sorted(sent_scores.items(), key = lambda x: x[1], reverse = True)[ : sent_length]]
    summary = ' '.join(summary)
    
    return summary
```

```python
summary_list = []
for i in range(len(df['article_content'])):
    # print(i)
    summary_list.append(summary(sent_scores(token_frequencies(i), i), 6))
```

# Sentiment analysis of article_summary

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
analyser = SentimentIntensityAnalyzer()
```

```python
df['article_title'][0]
```

```python
analyser.polarity_scores(df['article_summary'][0])
```

```python
analyser.polarity_scores(df['article_content'][0])
```

```python
def article_sentiment(article_content):
    return analyser.polarity_scores(article_content)['compound']

df['sentiment_score'] = df['article_content'].apply(article_sentiment)
```
